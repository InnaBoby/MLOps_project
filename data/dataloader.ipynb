```python
import kagglehub
from pathlib import Path
import re
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("ferraz/cotatenis-sneakers")

print("Path to dataset files:", path)
```

```python
def create_dataset(directory, output_file):
    """
    Scan directory to collect files, parse filenames to extract image metainformation (articul, brand, model), create and save pandas DataFrame

    Args:
        directory: source directory with pictures
        output_file: CSV filename with dataset info

    Returns:
        None
    """

    # Extract filenames from data directory
    files = [f.name for f in directory.iterdir() if f.is_file()]

    # Parsing filename to extract articul, brand and model
    pattern = r"([A-Za-z0-9-]+)_Tenis-([A-Za-z0-9]+)-([A-Za-z0-9-]+)\."

    filenames = []
    articuls = []
    brands = []
    models = []

    for file in files:
        match = re.match(pattern, file)

        if match:
            articul, brand, model = match.groups()

            filenames.append(file)
            articuls.append(articul)
            brands.append(brand)
            models.append(model)

    # Create DataFrame
    df = pd.DataFrame(
        {"filename": filenames, "articul": articuls, "brand": brands, "model": models}
    )

    # Save dataset to CSV
    df.to_csv(output_file, index=False)

    print(f"Датасет сохранён в {output_file}")
    return
```

```python
directory = Path(path, "sneakers_dataset/sneakers_dataset/")
output_file = "sneakers_dataset.csv"

create_dataset(directory, output_file)
```
